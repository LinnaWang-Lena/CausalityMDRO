```python
import csv
import numpy as np
import pandas as pd
import sys, os
import re
import ast
import datetime as dt
from tqdm import tqdm
from collections import defaultdict, Counter
import dill
import warnings

warnings.filterwarnings("ignore")
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
#MIMIC IV 24
chart_ts = pd.read_csv('/MIMIC_IV/Median_B/A24_Chart_median_max_nan.csv')
lab_ts = pd.read_csv('/MIMIC_IV/Median_B/A24_Lab_median_max_nan.csv')
proc_ts = pd.read_csv('/MIMIC_IV/Median_B/A24_Proc_median_max_nan.csv')
Meds_ts = pd.read_csv('/MIMIC_IV/Median_B/A24_Med_sum_nan.csv')
diags = pd.read_csv('/MIMIC_IV/Median_B/All_diags_binary.csv')
print(chart_ts.shape,Meds_ts.shape,lab_ts.shape,proc_ts.shape,diags.shape)
```

```python
f_icu = list(set(chart_ts.ICUSTAY_ID.unique())&set(lab_ts.ICUSTAY_ID.unique())&set(proc_ts.ICUSTAY_ID.unique())&set(Meds_ts.ICUSTAY_ID.unique()))
len(f_icu)
```

```python
chart_ts = chart_ts[chart_ts.ICUSTAY_ID.isin(f_icu)]
lab_ts = lab_ts[lab_ts.ICUSTAY_ID.isin(f_icu)]
proc_ts = proc_ts[proc_ts.ICUSTAY_ID.isin(f_icu)]
Meds_ts = Meds_ts[Meds_ts.ICUSTAY_ID.isin(f_icu)]
print(chart_ts.shape,Meds_ts.shape,lab_ts.shape,proc_ts.shape,diags.shape)
```

```python
all_ts = pd.merge(chart_ts,lab_ts,on='ICUSTAY_ID',how='left')
print(all_ts.shape)
all_ts = pd.merge(all_ts,proc_ts,on='ICUSTAY_ID',how='left')
print(all_ts.shape)
all_ts = pd.merge(all_ts,Meds_ts,on='ICUSTAY_ID',how='left')
print(all_ts.shape)
```

```python
all_ts.to_csv('/MIMIC_IV/Median_B/All_TS_non_49023.csv',index=False)
```
